# merge

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[34,5,6,7]})
var2 = pd.DataFrame({"A":[1,2,3,5],"c":[45,6,7,8]})


pd.merge(var1,var2,on="A") #frame for common data
```

```python
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","ID"))
```

```python
pd.merge(var1,var2,how="left") 
```

```python
pd.merge(var1,var2,how="right") 
```

```python
pd.merge(var1,var2,how="outer") 
```

```python
pd.merge(var1,var2,how="inner") 
```

```python
pd.merge(var1,var2,how="outer",indicator=True) 
```

concat

```python
sr1=pd.Series([1,2,3,4])
sr2=pd.Series([11,21,31,41])

pd.concat([sr1,sr2])
```

```python
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,23,45,33]})
d2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,5]})
pd.concat([d1,d2]) #all array must be in same length
pd.concat([d1,d2],axis=0) #rows
pd.concat([d1,d2],axis=1) #column
```

```python
pd.concat([d1,d2],axis=1,join="inner",keys=["d1","d2"])
```

```python
pd.concat([d1,d2],axis=0,join="inner",keys=["d1","d2"])
```

# join

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,2,3,23]})
var2=pd.DataFrame({"C":[10,20,30,40],"D":[11,22,33,44]})

var1.join(var2)
```

```python
var1.join(var2,how="left")
```

```python
var1.join(var2,how="right") #inner,outer,left
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,2,3,23]},index=["a","b","c","d"])
var2=pd.DataFrame({"C":[10,20],"B":[11,22]},index=["a","b"])

var1.join(var2,how="inner",lsuffix="_12",rsuffix="_12")
```

# Append

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,2,3,23]})
var2=pd.DataFrame({"C":[10,20],"B":[11,22]})

var1.append(var2,ignore_index=True)
```

# group_by

```python
import pandas as pd
var=pd.DataFrame({"name":["a","b","c","d","a","b","a","b","a","c","c"],
                  "s_1":[12,13,14,15,16,17,18,19,45,45,33],
                  "s_2":[12,13,14,15,16,17,18,19,45,22,45],
                             })
var
```

```python
var_new=var.groupby("name")
var_new
for x,y in var_new:
    print(x)
    print(y)
    print()
```

```python
var_new.get_group("a")
```

```python
var_new.min()
```

```python
var_new.max()
```

```python
lst=list(var_new)
lst
```

# pivot()

```python
import pandas as pd

var=pd.DataFrame({"days":[1,2,3,4,5,6],"stud_name":["a","b","c","a","b","c"],"eng":[10,11,13,12,12,43],"math":[10,11,13,12,12,43]})
var
```

```python
var.pivot(index="days",columns="stud_name")
```

```python
var.pivot_table(index="stud_name",columns="days",aggfunc="mean",margins=True)
```
